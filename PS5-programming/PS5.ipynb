```python
def E(h, beta, eta,x,y):
    xj = np.zeros_like(x)
    xj[:-1, :] = x[1:, :]  # down
    xj[1:, :] += x[:-1, :]  # up
    xj[:, :-1] += x[:, 1:]  # right
    xj[:, 1:] += x[:, :-1]  # left
    return h * np.sum(x) - beta * np.sum(xj * x)  - eta * np.sum(x*y)
def neighbors(x,i,j):
    if i == 0 and j == 0:
        neighbors = [[i,j+1],[i+1,j]]
    elif i == 0 and j == x.shape[1]-1:
        neighbors = [[i,j-1],[i+1,j]]
    elif i ==x.shape[0]-1 and j == 0:
        neighbors = [[i,j+1],[i-1,j]]
    elif i ==x.shape[0]-1 and j ==x.shape[1]-1:
        neighbors = [[i,j-1],[i-1,j]]
    elif j == 0:
        neighbors = [[i,j+1],[i+1,j],[i-1,j]]
    elif j == x.shape[1]-1:
        neighbors = [[i,j-1],[i+1,j],[i-1,j]]
    elif i == 0:
        neighbors = [[i,j+1],[i,j-1],[i+1,j]]
    elif i == x.shape[0]-1:
        neighbors = [[i,j+1],[i,j-1],[i-1,j]]

    else:
        neighbors = [[i,j-1],[i-1,j],[i,j+1],[i+1,j]]
    return neighbors
def calc_new_E(h,beta,eta,E,i,j,x,y):
    old = x[i,j]
    new = -1*old
    E_new = E - (h * old) + (h * new) + (eta * y[i, j] * old) - (eta * y[i, j] * new)
    neighbor = neighbors(x,i,j)
    E_new = E_new + 2* beta * sum(x[a[0],a[1]] * old for a in neighbor)- 2 * beta * sum(x[a[0],a[1]] * new for a in neighbor)
    
    return E_new
```

```python
#### insert your implementation here
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib._png as png
import numpy.matlib

###############################
# 0. Set the parameter numbers#
###############################
# number of nodes
N = 40000

# number of states
K = 32

# the coefficients for data term
lambda_data = 0.04

# the coefficients for the smoothness term
lambda_smooth = 0.04

# the potencial function
m_potential_func = 'L2'

raw_img = png.read_png_int("Lena-noise.png")
data=raw_img.reshape(1,N)/255


```

```python
##################################
# 1. Setup the potential function#
##################################

# define uniary potentials
if m_potential_func == 'L1':
    P = -lambda_data * abs(np.matlib.repmat(data,K,1) - np.matlib.repmat(np.arange(1,K+1,1).reshape((K,1)),1,N))

elif m_potential_func == 'L2':
    P = -lambda_data * (np.matlib.repmat(data,K,1) - np.matlib.repmat(np.arange(1,K+1,1).reshape((K,1)),1,N))**2

print(P.shape, P[:,0].shape, P[:,0].reshape(K,1).shape, P[:,0])

foo = np.matlib.repmat(np.arange(1,K+1,1),K,1)
dist = abs(foo - np.transpose(foo))

psi = {}
print()
# define pair-wise potentials
for i in range(N-1):
    if m_potential_func == 'L1':
        psi[i] = -lambda_smooth*dist
    elif m_potential_func == 'L2':  
        psi[i] = -lambda_smooth*dist**2;

# compose the potentials
for i in range(N-1):
    #psi[i] = np.matlib.repmat(P[:,i],1,K) + psi[i]
    psi[i] = np.matlib.repmat(P[:,i].reshape(K,1),1,K) + psi[i]
psi[N-2] = np.matlib.repmat(np.transpose(P[:,N-1]),K,1) + psi[N-2]

print(psi[N-2].shape, psi[N-2])
```

```python
######################
# 2. Message passing #
######################
# forward and backward messages
fm = {}
bm = {}
for i in range(N):
    fm[i] = np.zeros((1,K))
    bm[i] = np.zeros((K,1))
    
# energy
fm[0] = np.log(np.sum(np.exp(psi[0]),0))
bm[N-1] = np.log(np.sum(np.exp(psi[N-2]),1))
fm0 = fm.copy()
bm0 = bm.copy()
# forward message passing
for i in range(1, N-1):
    fm[i] = np.log(np.sum(np.exp(np.matlib.repmat(np.transpose(fm0[i-1]).reshape(K,1),1,K) + psi[i]), 0))
    fm[i] = fm[i] - np.min(fm[i])
    
# backward message

for i in range(N-2,1,-1):
    x = np.exp(np.matlib.repmat(np.transpose(bm0[i+1]),K,1) + psi[i-1])
    x = np.sum(x, 1)
    bm[i] = np.log(x)
    bm[i] = bm[i] - np.min(bm[i])
# fm0 = fm.copy()
# bm0 = bm.copy()
print(fm[0].shape, fm[1].shape)
print(fm[0], fm[1])
```

```python
########################
# 3. Compute marginals #
########################
marginal = {}
for i in range(N):
    marginal[i] = np.zeros((1,K))

for i in range(N):
    if i > 0:
        x = (marginal[i] + fm[i-1])
        marginal[i] = marginal[i] + fm[i-1]
    if i < N-1:
        marginal[i] = marginal[i] + np.transpose(bm[i+1])
    marginal[i] = np.exp(marginal[i] - np.min(marginal[i]))
    # normalize the marginal
    marginal[i] = marginal[i]/ np.sum(marginal[i])
    
Marginals = np.zeros((K,N))
DataOnly = np.zeros((K,N))

for i in range(N):
    Marginals[:,i] = marginal[i].reshape(K)
    DataOnly[:,i] = np.exp(P[:,i]).reshape(K)
    DataOnly[:,i] = DataOnly[:,i] / np.sum(DataOnly[:,i])

print(marginal[0].shape)
```
